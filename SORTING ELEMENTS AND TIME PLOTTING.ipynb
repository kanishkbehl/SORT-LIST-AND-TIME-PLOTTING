# Q12.Sort following number of elements in a list, calculate time taken and plot the graph.

50K Elements

```python
import random as r
import timeit as t
import matplotlib.pyplot as plt

L = []
for i in range(50000):
    otp = r.randint(0, 1000000)
    L.append(otp)

start = t.default_timer()
L.sort()
stop = t.default_timer()
T1 = stop - start
print("T1 =", T1)
```

100K Elements

```python
import random as r
import timeit as t
L=[0]
for i in range(0,100000):
  otp=r.randint(0,1000000)
  L.append(int(otp))
start=t.default_timer()
L.sort()
stop=t.default_timer()
T2=stop-start
print("T2=",T2)
```
